```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

path = "https://drive.google.com/uc?export=download&id=1UwpW8GELggsVNj1IHtAo5rm4js_6b6PZ"
data = pd.read_csv(path)
x = data.drop('variety', axis=1)
y = data['variety']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

"""
mlp = MLPClassifier(hidden_layer_sizes=(15, 10), max_iter=1000, random_state=42)
mlp.fit(x_train, y_train)
y_pred = mlp.predict(x_test)
"""

from tensorflow.keras.utils import to_categorical

y_train_onehot = pd.get_dummies(y_train).values
y_test_onehot = pd.get_dummies(y_test).values

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create neural network model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(4,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(x_train, y_train_onehot, epochs=100, batch_size=32, verbose=0)

y_pred=model.predict(x_test)
y_pred_class = np.argmax(y_pred, axis=1)
accuracy = accuracy_score(np.argmax(y_test_onehot, axis=1), y_pred_class)

print(f"Accuracy: {accuracy}")
```
